```python
df_hr = pd.read_csv('/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/heartrate_seconds_merged.csv')
```

```python
df_hr.info()
df_hr.shape
```

```python
df_hr['Time'] = pd.to_datetime(
    df_hr['Time'],
    format="%m/%d/%Y %I:%M:%S %p",
    errors='coerce')
```

```python
df_hr['ActivityMinute'] = df_hr['Time'].dt.floor('min')
df_hr.drop(columns='Time', inplace=True)
```

```python
df_hr = (
    df_hr
    .groupby(['Id', 'ActivityMinute'], as_index=False)['Value']
    .mean()
    .rename(columns={'Value': 'AvgHeartRate'})
)
```

```python
df_hr['AvgHeartRate'] = df_hr['AvgHeartRate'].round().astype(int)
```

```python
df_hr.info()
```

**Note:**
Heart rate measurements (`AvgHeartRate`) appear to be incomplete — not every `Id` and `ActivityMinute` pair has a corresponding value.
To avoid introducing `NaN` values during merging, I’ve chosen **not to join** the heart rate table at this stage.
Further analysis will be performed separately, with attention to data coverage and consistency.

### Edited

While working with visualization in Tableau, I noticed a problem: the data on daily activity does not quite match the daily data on sleep. That is, daily activity occurs on a specific day, while sleep usually occurs on the following day (00:00 - 23:59). Therefore, I will make an adjustment so that it is possible to aggregate correctly by day.

```python
  df_sleep['activity_date'] = df_sleep['SleepDay'] - pd.Timedelta(days=1)
```
